Import libraries

```python
# requests to call the API
import requests 

# JSON to format the data
import json
```

URL for the CSO API endpoint returning the FIQ02 dataset in JSON

```python
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/1.0/en"
```

Create function to request the dataset and return the JSOM response

```python
# Sends a GET request to the API and return the response as JSON

def getAll():
    response = requests.get(url)
    return response.json()
```

```python
# Write the JSON data to a file and create a new file called cso.json

if __name__ == "__main__":
    with open("cso.json", "wt") as fp:
        print(json.dumps(getAll()), file=fp)
```
