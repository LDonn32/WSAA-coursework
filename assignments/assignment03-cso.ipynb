Import libraries

```python
import requests 
import json
```

URL for the CSO API endpoint returning the FIQ02 dataset in JSON

```python
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/1.0/en"
```

Create function to request the dataset and return the JSOM response

```python
def getAll():
    response = requests.get(url)
    return response.json()
```

```python
if __name__ == "__main__":
    with open("cso.json", "wt") as fp:
        print(json.dumps(getAll()), file=fp)
```
